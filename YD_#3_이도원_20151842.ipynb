### 1) 팀별  성공률(드리블,  공격지역 패스, 크로스)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
# backend를 inline으로 설정
%matplotlib inline
```

```python
data.rename(columns={'공격지역패스성공률': '공격지역패스', '드리블성공률': '드리블', '크로스성공률': '크로스'}, inplace=True)
data1 = data.pivot_table(index = '팀', aggfunc = 'mean', values = ['드리블', '공격지역패스', '크로스'])
data1 = data1.stack(0)
data1 = data1.unstack(0)
data1
```

```python
data1.plot(kind = 'line', # 차트의 종류
       title = '팀별 성공률', # 그래프 제목
        figsize = (12,12), # figure의 크기
        lw = 2, # line width
        subplots = False, # 기본값 = False -> 하나의 axes에 모든 데이터 표현.  True로 설정하면, 계열 별로 axes를 생성함.
        ylim = [5, 70] # y축의 min,max 설정
       )
plt.ylabel('성공률')
plt.axis('on')
plt.show()
```

### 4) 공격수 중 출전시간이 적은 사람과 많은 사람 차이점 분석

```python
data.rename(columns={'출전시간(분)': '출전시간'}, inplace=True)
data['출전시간'].mean()
```

```python
sub1 = data[(data.포지션 == 'FW') & (data.출전시간>447)]
sub1['유효슈팅률'] = sub1['유효 슈팅'] / sub1['슈팅'] * 100
sub1 = sub1.dropna(axis=0)
sub1 = sub1.mean()
sub1 = sub1.to_frame(name='출전다')
sub1 = sub1.transpose()
sub1 = sub1[['드리블','공격지역패스','크로스','유효슈팅률']]
sub1
```

```python
sub2 = data[(data.포지션 == 'FW') & (data.출전시간<=447)]
sub2['유효슈팅률'] = sub2['유효 슈팅'] / sub2['슈팅'] * 100
sub2 = sub2.dropna(axis=0)
sub2 = sub2.mean()
sub2 = sub2.to_frame(name='출전소')
sub2 = sub2.transpose()
sub2 = sub2[['드리블','공격지역패스','크로스','유효슈팅률']]
sub2
```

```python
data4 = pd.concat([sub1,sub2])
data4 = data4.transpose()
data4
```

```python
data4.plot(kind = 'bar', figsize = (10,10))
```

### 5) 출전 시간 대비 가장 실수를 많이 하는 공격수 Best5

```python
data5 = data[data.포지션=='FW']
data5['실수빈도'] = data5['볼미스'] / data5['출전시간']
data5 = data5.dropna(axis=0)
data5 = data5.sort_values(by = '실수빈도', ascending = False)[:5]
data5 = data5[['선수명','실수빈도']]
data5['실수퍼센트'] = data5['실수빈도']/data5['실수빈도'].sum()*100
data5
```

```python
ratio = data5['실수퍼센트']
labels = data5['선수명']
plt.pie(ratio, labels=labels, autopct='%.1f%%')
plt.show()
```
